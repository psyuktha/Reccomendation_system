```python
# prompt: merge ratings and movies.csv based on movieid

movies = pd.read_csv('/content/drive/MyDrive/ml-latest-small/movies.csv')
ratings = pd.read_csv('/content/drive/MyDrive/ml-latest-small/ratings.csv')

merged_df = pd.merge(movies, ratings, on='movieId')
```

```python
merged_df
```

```python
# prompt: make data frame with movie id title and rating

df = merged_df[['userId','movieId', 'title', 'rating']]
```

```python
df.head() ,len(df)
```

```python
rating_count=df.groupby(by='title')['rating'].count()
rating_count=rating_count.reset_index()
rating_count=rating_count.rename(columns={'rating':'tot_rate_count'})
rating_count
```

```python
final_rating=pd.merge(df,rating_count,on='title')
final_rating
```

```python
# prompt: describe the table

final_rating.describe()
```

```python
threshold=50
```

```python
final_rating = final_rating[final_rating['tot_rate_count'] > threshold]
final_rating
```

```python
# prompt: create pivot matrix

final_rating_pivot = final_rating.pivot_table(columns='userId', index='title', values='rating')
final_rating_pivot
```

```python
# prompt: replace NaN with 0.0

final_rating_pivot.fillna(0.0, inplace=True)
final_rating_pivot
```

```python
from scipy.sparse import csr_matrix

movie_features_df_matrix = csr_matrix(final_rating_pivot.values)
from sklearn.neighbors import NearestNeighbors


model_knn = NearestNeighbors(metric = 'cosine', algorithm = 'brute')
model_knn.fit(movie_features_df_matrix)
```

```python
# query_index = np.random.choice(final_rating_pivot.shape[0])
# print(query_index)
distances, indices = model_knn.kneighbors(final_rating_pivot.iloc[108,:].values.reshape(1, -1), n_neighbors = 6)
movie_index = final_rating_pivot.index.get_loc('Crimson Tide (1995)')
print(movie_index)
```

```python
distances,indices
```

```python
for i in range(0, len(distances.flatten())):
    if i == 0:
        print('Recommendations for {0}:\n'.format(final_rating_pivot.index[108]))
    else:
        print('{0}: {1}, with distance of {2}:'.format(i, final_rating_pivot.index[indices.flatten()[i]], distances.flatten()[i]))
```
